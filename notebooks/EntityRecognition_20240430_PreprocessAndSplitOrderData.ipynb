## Import library

```python
import json
import random
```

## Read data

```python
with open('../data/labeled/entity/order/human_data_full.json', 'r', encoding='utf-8') as file:
    human_data = json.load(file)

with open('../data/labeled/entity/order/chatgpt_data.json', 'r', encoding='utf-8') as file:
    chatgpt_data = json.load(file)
```

## Split data into 3 sets

Include:
1. Data which is 100% generated by ChatGPT 3.5
2. Data which is 100% written by human
3. Combined data which is aggregated from 2 previous sources in 50-50 ratio

```python
case_1_train = random.sample(chatgpt_data, 700)
case_1_train_ids = {id(item) for item in case_1_train}
test_gpt = random.sample([item for item in chatgpt_data if id(item) not in case_1_train_ids], 300)
test_gpt_ids = {id(item) for item in test_gpt}

case_3_train = random.sample(human_data, 700)
case_3_train_ids = {id(item) for item in case_3_train}
test_human = random.sample([item for item in human_data if id(item) not in case_3_train_ids], 300)
test_human_ids = {id(item) for item in test_human}

case_2_gpt = random.sample([item for item in chatgpt_data if id(item) not in test_gpt_ids], 350)
case_2_human = random.sample([item for item in human_data if id(item) not in test_human_ids], 350)
```

```python
with open("case_1_train.json", 'w', encoding='utf-8') as f:
    json.dump(case_1_train, f, ensure_ascii=False, indent=4)
```
